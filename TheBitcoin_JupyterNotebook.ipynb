# Should you be tempted to invest in Bitcoin? 

Bitcoin, the new hot currency created in 2009 by a person with alias Satoshi Nakamoto. No banks are required for the transactions plus no transaction fees and no need to reveal your identity while trading. These are some features that make it unique and happening but the biggest question is **Will it survive?** People have their beliefs by observing its super fluctuating price trends, that it can crash anytime. Also, there is no legal regulation on it like other currencies but it does provide cheaper transactions with ease of use. <br />
To reach a decision, whether it will be a good idea to invest in the bitcoin or not? we will be analyzing some key metrics about the market trends. We have multiple data sets web scraped and downloaded from websites like kaggle.com, quandl.com, coindesk.com, detailed data sources are mentioned below. <br />  **Why we selected this data?** Because the audience we are presuming are "Investors looking to diversify their investment portfolio". So, while investing somewhere they would certainly like to know how this particular currency is better than the others in the market and also would like to look at the historical price trends of each to make a decision. The details about the datasets is mentioned along with the data source links. After getting the data sets, we integrated the datasets as per the common attributes and got 3 different datasets in the end. <br />

#### Merging the Bitcoin, Litecoin, Monero data into one dataset, called Dataset1

```python
Dataset1 = Bitcoin_litecoin.merge(Monero, on="Date", how="outer")
```

#### Displaying row number 120 - 125 from the integrated dataset

```python
Dataset1.iloc[120:125]
```

#### Renaming the column names for Dataset1

```python
Dataset1 = Dataset1.rename(columns={'Open': 'Monero_Open', 
                              'High': 'Monero_High', 
                              'Low':'Monero_Low', 
                              'Close': 'Monero_Close',
                              'Volume': 'Monero_Volume', 
                              'Market Cap': 'Monero_Market Cap',
                              })
```

#### Now, merging the above merged data set with Ethereum prices

```python
Dataset1 = Dataset1.merge(Ethereum, on="Date", how="outer")
```

```python
Dataset1.head()
```

#### Renaming the column names for Ethereum

```python
Dataset1 = Dataset1.rename(columns={'Open': 'Ethereum_Open', 
                              'High': 'Ethereum_High', 
                              'Low':'Ethereum_Low', 
                              'Close': 'Ethereum_Close',   
                              'Volume': 'Ethereum_Volume', 
                              'Market Cap': 'Ethereum_Market Cap',
                              })
```

#### Merging the Dataset1 with S&P 500 Index based on common attribute 'Date'

```python
Dataset1 = Dataset1.merge(SP500, on="Date", how="outer")
```

#### Again merging the integrated dataset with Nasdaq Index data on Bitcoin

#### The next data is about the number of transactions happening in one day for bitcoin and Ethereum, merging with EURO and USD dollar stock prices against Bitcoin
